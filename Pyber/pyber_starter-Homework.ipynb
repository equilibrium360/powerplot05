```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"


# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset

city_ride_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])
# Display the data table for preview
city_ride_data.head(10)
```

## Bubble Plot of Ride Sharing Data

```python
urban_data =  city_ride_data.loc[city_ride_data["type"] == "Urban"]
suburban_data =  city_ride_data[city_ride_data["type"] == "Suburban"]
rural_data =  city_ride_data[city_ride_data["type"] == "Rural"]


urban_ride_count = urban_data.groupby("city")["ride_id"].count()
urban_driver_count = urban_data.groupby("city")["driver_count"].sum()
urban_avg_fare = urban_data.groupby("city")["fare"].mean()

suburban_ride_count = suburban_data.groupby("city")["ride_id"].count()
suburban_driver_count = suburban_data.groupby("city")["driver_count"].sum()
suburban_avg_fare = suburban_data.groupby("city")["fare"].mean()

rural_ride_count = rural_data.groupby("city")["ride_id"].count()
rural_driver_count = rural_data.groupby("city")["driver_count"].sum()
rural_avg_fare = rural_data.groupby("city")["fare"].mean()


plt.scatter(urban_ride_count, urban_avg_fare, s=urban_driver_count, c="orange", alpha=0.70)
plt.scatter(suburban_ride_count, suburban_avg_fare, s=suburban_driver_count, c="blue", alpha=0.70)
plt.scatter(rural_ride_count, rural_avg_fare, s=rural_driver_count, c="green", alpha=0.70)
 # Add labels to the x and y axes
plt.title("Pyber Ride sharing Data (2016)")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare ($) Per City")

plt.grid(True)

plt.show()




```
